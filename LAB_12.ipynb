```python
# STEP 5: Load the data into pandas
import pandas as pd

df = pd.read_csv("data/Mall_Customers.csv")
print("Train data shape:", df.shape)
print("\nTrain data info:") # New line for better formatting
df.info() # Call info directly
print("\nTrain data statistics:") # New line for better formatting
df.describe() # Call describe directly
```

```python
# Label Encoding of gender
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df.head()
```

```python
# min max sclar on "Age , spending score , annual income ""
from sklearn.preprocessing import MinMaxScaler
variables = ["Age","Annual Income (k$)","Spending Score (1-100)"]
scaler = MinMaxScaler()
df[variables] = scaler.fit_transform(df[variables])

df.head() # Call info directly
```

```python
# kmeans clusturing "5"
from sklearn.cluster import KMeans
cluster = 5
kmeans = KMeans(n_clusters=cluster, random_state=42)
kmeans.fit(df[variables])

cluster_labels = kmeans.labels_
df["Cluster"] = cluster_labels
df.head()
```
