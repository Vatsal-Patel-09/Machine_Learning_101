### Draw scatter plot for
#### 1: mpg vs cylnders (Scatter plot)
#### 2: histogram for "mpg"

```
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"]
```

```
from matplotlib import pyplot as plt
import pandas as pd
```

```
df = pd.read_csv(data_url, names=column_names, delim_whitespace=True, na_values="?")
df.head(10)
```

```
plt.scatter(df["mpg"], df["cylinders"])
plt.xlabel("MPG")
plt.ylabel("Cylinders")
plt.title("Scatter plot of MPG vs Cylinders")
plt.show()
```

```
plt.hist(df["mpg"])
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.title("Histogram of MPG")
plt.show()
```

### Irise data set plot scatter plot between petal width vs sepal width

```
data_url_iris = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```
column_names_iris = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
```

```
df_iris = pd.read_csv(data_url_iris, names = column_names_iris,delimiter="," ,na_values="?")
df_iris.head(10)
```

```
plt.scatter(df_iris["petal_width"], df_iris["sepal_width"])
plt.xlabel("petal width")
plt.ylabel("sepal width")
plt.title("Scatter plot of petal width vs sepal width")
plt.show()
```
